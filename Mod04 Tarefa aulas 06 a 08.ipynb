####  2) Melhorando a interação com o API
Observe dois URLs de consultas diferentes, por exemplo o URL utilizado em aula, e o URL feito no exercício anterior. Compare-os e observe as diferenças.

1. Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado (```'Estado'```ou ```Município```), e que devolve os dados da consulta no formato *DataFrame*.
2. Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com *status* em 'Arquivado por decurso de prazo' estão registradas?
3. Qual é o município da Bahia com mais solicitações deferidas?
4. Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'

````python
#1) Seu código aqui

def consultar_dados_uf_tipo_interessado(uf, tipo_interessado):
    
    url = f"http://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={uf}&tipo_interessado={tipo_interessado}"
    
    
    r = requests.get(url)
    
    
    if r.status_code == 200:
        
        data_json = r.json()
        
        
        df = pd.DataFrame(data_json['items'])
        
        return df
    else:
        print(f"Erro ao acessar os dados: {r.status_code}")
        return None

uf = 'SP'
tipo_interessado = 'Estado'

dados = consultar_dados_uf_tipo_interessado(uf, tipo_interessado)

print(dados.head())
````

````python
# 2) Seu código aqui

dados_mg = consultar_dados_uf_tipo_interessado('MG', 'Estado')

solic_arquivadas_mg = dados_mg[dados_mg['status'] == 'Arquivado por decurso de prazo']

num_solic_arquivadas_mg = len(solic_arquivadas_mg)

print(f"Número de solicitações para o Estado de Minas Gerais com status 'Arquivado por decurso de prazo': {num_solic_arquivadas_mg}")
````

````python
# 3) Seu código aqui

def municipio_com_mais_solicitudes_deferidas(uf):
    
    dados_uf = consultar_dados_uf_tipo_interessado(uf, 'Municipio')
        
    solic_deferidas = dados_uf[dados_uf['status'] == 'Deferido']
        
    contagem_solic_deferidas = solic_deferidas.groupby('municipio').size()
        
    municipio_mais_solic_deferidas = contagem_solic_deferidas.idxmax()
        
    return municipio_mais_solic_deferidas

municipio_mais_solic_deferidas_bahia = municipio_com_mais_solicitudes_deferidas('BA')

print(f"O município da Bahia com mais solicitações deferidas é: {municipio_mais_solic_deferidas_bahia}")
````

````python
# 4) Seu código aqui

uf = 'BA'
tipo_interessado = 'Estado'

dados_bahia = consultar_dados_uf_tipo_interessado(uf, tipo_interessado)


if dados_bahia is not None:
    
    nome_arquivo = 'solicitacoes_bahia_estado.csv'
    dados_bahia.to_csv(nome_arquivo, index=False)
    print(f"Os dados foram salvos com sucesso no arquivo: {nome_arquivo}")
else:
    print("Não foi possível obter os dados da Bahia com interessado = 'Estado'.")
````
